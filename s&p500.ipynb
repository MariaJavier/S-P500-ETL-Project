```python
# Print the json to view it
pprint(json_data)
```

We transformed the json into a data frame

```python
ct = 0
jname = []
jsector = []
jsymbol = []

for x in json_data:
    jname.append(json_data[ct]["Name"])
    jsector.append(json_data[ct]["Sector"])
    jsymbol.append(json_data[ct]["Symbol"])
    ct = ct +1
    
df = pd.DataFrame({
    "Name":jname,
    "Sector":jsector,
    "Symbol":jsymbol
}
)
df.head()
```

We are making three lists below in a for loop. These three list will be used in making a data frame.

```python
percent_list = []
open_list = []
close_list = []
ct = 0
for x in list(range(0,505)):
    
    percent_list.append(100*(list_list[ct].close.mean()-list_list[ct].open.mean())/list_list[ct].open.mean())
    open_list.append(list_list[ct].open.mean())
    close_list.append(list_list[ct].close.mean())
    ct = ct + 1
```

This is the second data frame we have. it will be merged later. 

```python
df2 = pd.DataFrame({
    "Name":jname,
    "Symbol":jsymbol,
  
    "2013 Open":open_list,
    "2018 Close":close_list
})
df2
```
